```python
#url of the site on which web scrapping needs to be done
url = "https://www.flipkart.com/mobiles/pr?sid=tyy%2F4io&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.availability%255B%255D%3DExclude%2BOut%2Bof%2BStock&otracker=clp_metro_expandable_1_apple-categ-94e9d_iPhone_apple-products-store_90ff40fd-a46b-4a40-9440-fbe783136afb_DesktopSite_wp1&fm=neo/merchandising&iid=M_5a9223de-a054-43a2-a265-1882411856c7_1.90ff40fd-a46b-4a40-9440-fbe783136afb_DesktopSite&ppt=CLP&ppn=CLP:pixel-2-1514-0309-stor"
url
```

```python
#Checking the response status of the url
def response_url(site):
    http = urllib3.PoolManager()
    response = http.request('GET',site)
    return response.status

response_url(url)
```

```python
uClient = uReq(url)
page_html= uClient.read()
uClient.close()
```

```python
page_soup = soup(page_html, "html.parser")
page_soup.h1
```

```python
item_container= page_soup.find_all("div",{"class":"col _2-gKeQ"})
item_container
```

```python
abc=item_container[0].find_all("li",{"class":"tVe95H"})
abc
```

```python
#Stroing the information of description into a list
desc=[abc[0].text,abc[1].text,abc[2].text,abc[3].text,abc[4].text,abc[5].text]
ram=abc[0].text
display=abc[1].text
extra=abc[2].text
extra1=abc[3].text
extra2=abc[4].text.replace(",","|")
extra3=abc[5].text.replace(",","|")
```

```python
#writing the data into a csv
filename="products.csv"
f=open(filename,"w")

headers="brand, url , desc, price, emi, time ,date, Status\n"
f.write(headers)
for container in item_container:
    brand=container.div.a.img["alt"].replace(",","|")
    print("Brand: " +brand)
    
    linkurl='https://www.flipkart.com'
    link=container.div.a["href"]
    fulllink=linkurl+link
    print("wrl: "+fulllink)
    
    abc=container.div.a.ul.find_all("li",{"class":"tVe95H"})
    desc=[abc[0].text,abc[1].text,abc[2].text,abc[3].text,abc[4].text,abc[5].text]
    print("Description: ")
    print(ram)
    print(display)
    print(extra)
    print(extra1)
    print(extra2)
    print(extra3)
    
    offer_price=container.div.a.find_all("div",{"class":"_1vC4OE _2rQ-NK"})
    offer=offer_price[0].text.replace("\u20b9","").replace(",","")
    print("Price: "+offer)
    
    emi=container.div.a.find_all("div",{"class":"_3MCpsc"})
    pemi=emi[0].text.replace("\u20b9","|").replace("EMI starting from |","")
    print("Emi: " +pemi)
    
    dt= datetime.datetime.now().date()
    tm=datetime.datetime.now()
    currtime=str(tm)
    currdate= str(dt)
    
    r= str(response_url(fulllink))
    print("Response Status: " + r)
    print("\n")
    
    
    f.write(brand+","+ fulllink +","+ram+" "+display+" "+extra+" "+extra1+" "+extra2+" "+ extra3+","+ offer +","+pemi+","+currtime+","+currdate+","+r +"\n")
f.close() 

```
